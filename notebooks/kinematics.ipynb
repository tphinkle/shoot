# Kinematics

## Imports

```python
# Python standard library

# Scientific computing
import numpy as np
import matplotlib.pyplot as plt
```

## Jumping

#### Acceleration based

```python
factor = 6400.
jumping_acceleration = -.5*factor
gravity_acceleration = .25*factor
acceleration_time = .2

physics_fps = 240.
physics_period = 1./physics_fps


num_updates = 200



ts = []
ys = []
vs = []
ays = []
vy = 0
y = 0


for i in range(num_updates):
    time = i*physics_period
    if time < acceleration_time:
        acceleration = gravity_acceleration + jumping_acceleration
    else:
        acceleration = gravity_acceleration
        
    
        
    vy += acceleration*physics_period
    
    y += vy*physics_period
    
    ts.append(time)
    ys.append(y)
    vs.append(vy)
    ays.append(acceleration)
    
ts = np.array(ts)
ys = np.array(ys)
vs = np.array(vs)
ays = np.array(ays)
    
    
plt.plot(ts, ys*(-1))
plt.show()

plt.plot(ts, vs*(-1))
plt.show()

plt.plot(ts, ays*(-1))
plt.show()
```
